```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import base64
```

```python
import pandas as pd
df = pd.read_csv('zipcodeData.txt', delimiter = "\t")
df=df.drop(columns="Index")
df.replace("---", float("NaN"), inplace=True)
df=df.dropna()

df[['Income']] = df[['Income']].replace('[\$,]','',regex=True).astype(float)
df=df.sort_values(by = "Income" ,ascending =False)
df
```

```python
df.head(5).plot.bar(x="City",y="Income",color="green",rot=90)
plt.title("Highest Income Cities")
plt.plot()
```

```python
df.tail(5).plot.bar(x="City",y="Income",color="brown",rot=90)
plt.title("Lowest Income Cities")
plt.plot()
```
